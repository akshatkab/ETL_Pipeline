# Data Loading and Reading their Schema 

```
### I Have uploaded all the zip file in DBFS by using upload data option ,as I dont have free tier account in AWS 
files = dbutils.fs.ls("/FileStore/tables/")
for file in files:
    print(file.path)
```

### Here Above list of all the files for fact and dimension tables
So I'm extracting their schema and data type

Dimension Table schema

```
files = [
    "hier_clnd_dlm.gz",
    "hier_hldy_dlm.gz",
    "hier_invloc_dlm.gz",
    "hier_invstatus_dlm.gz",
    "hier_possite_dlm.gz",
    "hier_pricestate_dlm.gz",
    "hier_prod_dlm.gz",
    "hier_rtlloc_dlm.gz"
]

for i in files:

    df = spark.read \
    .options(header="True", inferSchema="True",delimiter='|') \
    .csv("dbfs:/FileStore/tables/{i}".format(i=i))
    print('Schemas for  file :- {i}'.format(i=i))
    df.printSchema()
```

Fact Table

```
files = files = [
    'fact_averagecosts_dlm.gz',
    'fact_transactions_dlm.gz'
]

for i in files:

    df = spark.read \
    .options(header="True", inferSchema="True",delimiter='|') \
    .csv("dbfs:/FileStore/tables/{i}".format(i=i))
    print('Schemas for  file :- {i}'.format(i=i))
    df.printSchema()
```
